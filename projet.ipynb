```python
df.columns=df.columns.str.replace(" ","_")
df.columns
```

Identifier et gérer les valeurs manquantes

```python

df.isnull().sum()
```

```python
df= df.dropna(subset="Postal_Code")
```

```python
df.isnull().sum()
```

Supprimer ou corriger les doublons

```python

df.drop_duplicates()
```

```python
df.dropna()
```

Corriger les types de colonnes (dates, numériques, catégories)

```python
df.dtypes
```

```python

df["Order_Date"] = pd.to_datetime(df["Order_Date"],format="%d/%m/%Y")
df["Ship_Date"] = pd.to_datetime(df["Ship_Date"],format="%d/%m/%Y")
df.dtypes
```

```python
df["Postal_Code"]= df["Postal_Code"].astype(int)
df.dtypes
```

Transformation et feature engineering

Extraire des informations temporelles : année, mois, trimestre

```python
df["Année"]= df["Ship_Date"].dt.year
df["Mois"]= df["Ship_Date"].dt.month
df["Trimestre"]= df["Ship_Date"].dt.quarter
df
```

```python
df.columns
```

Créer de nouvelles métriques : marge, taux de profit, ratio quantité / ventes

```python
df["Cost"] = df["Sales"]*0.6
df
```

```python
df["Marge"]= df["Sales"]-df["Cost"]
df["Marge"]
```

```python
df["Taux_Profit"] = df["Marge"] / df["Sales"] * 100
df["Taux_Profit"]
```

Catégoriser certaines colonnes (ex. segmentation clients, catégories de produits)

```python
def seg_client(valeur):
    if valeur>500:
        return "high"
    elif valeur >250:
        return "medium"
    else:
        return "low"
df["segment_client"] = df["Sales"].apply(seg_client)
df
```

Ajouter des colonnes dérivées pour analyse future (ex. délais livraison, indicateurs de performance)

```python
df["délais_livraison"]= df["Ship_Date"]-df["Order_Date"]
df
```

 indicateurs de performance

```python
def Performance(day):
    if day.days <=3:
        return "in time"
    else:
        return "late"
df["Performance"]= df["délais_livraison"].apply(Performance)
df
```

Agréger certaines données pour faciliter l’analyse (ex. ventes par produit ou région)

```python
ventes_produit = df.groupby("Product_Name")["Sales"].sum().round(2)
ventes_produit
```

Exporter le dataset propre

```python
df_final = df.copy()
df_final.to_csv("superstore_clean.csv", index=False)
```
